# K-Folder Validation

```
from keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```
from __future__ import print_function
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```
batch_size = 32
num_classes = 10
epochs = 30
```

```
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

```
from random import randrange

# Split x_train and y_train into k folds
def cross_validation_split(x_train, y_train, folds=3):
    x_train_split = list()
    y_train_split = list()
    x_train_copy = list(x_train)
    y_train_copy = list(y_train)
    
    fold_size = int(len(x_train) / folds)
    for i in range(folds):
        x_fold = list()
        y_fold = list()
        while len(x_fold) < fold_size:
            index = randrange(len(x_train_copy))
            x_fold.append(x_train_copy.pop(index))
            y_fold.append(y_train_copy.pop(index))
            
        x_train_split.append(x_fold)
        y_train_split.append(y_fold)
    return x_train_split, y_train_split
```

```
x,y = cross_validation_split(x_train, y_train, 5)
```

```
model = Sequential()
model.add(Conv2D(32, (3, 3), padding='same',
                 input_shape=x_train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.1))

model.add(Conv2D(32, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.1))

model.add(Conv2D(32, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.1))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.2))

model.add(Dense(num_classes))
model.add(Activation('softmax'))
```

```
from keras.optimizers import Adam
adam = Adam()
```

```
model.compile(loss='categorical_crossentropy',
              optimizer=adam,
              metrics=['accuracy'])
```

```
import numpy as np

for i in range(5):
    x_val = np.array(x[i])
    y_val = np.array(y[i])
    index = [0,1,2,3,4]
    index.remove(i)
    x_train = np.empty((0,32,32,3))
    y_train = np.empty((0,1))
    for j in index:
        x_train = np.append(x_train, x[j], axis=0)
        y_train = np.append(y_train, y[j], axis=0)
    
    y_train = keras.utils.to_categorical(y_train, num_classes)
    y_val = keras.utils.to_categorical(y_val, num_classes)
    
    print('------Traning Process of {} Folder'.format(i+1))
    
    history = model.fit(x_train, y_train,
                     batch_size=batch_size,
                     epochs=epochs,
                     validation_data=(x_val, y_val),
                     shuffle=True)
```
